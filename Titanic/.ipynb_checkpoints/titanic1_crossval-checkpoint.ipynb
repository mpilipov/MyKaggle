```python
import pandas as pd
import numpy as np

df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
df_s = pd.read_csv("gender_submission.csv")
```

```python
# 1. Нужно заменить Age на медианное значение, Cabin можно вычеркнуть
median_age = df_train.Age.median()
df_train['Age']=df_train['Age'].fillna(median_age)
df_train.head(20)
# 2. Также убираем как не представляющие ценности: PassengerId, Name, Ticket, Cabin
df_train=df_train.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
```

```python
df_train.info()
```

```python
# Что такое признаки Parch и SibSp?
# SibSp - количество братьев\сестёр
# Parch - количество родителей\детей на борту
print df_train['Embarked'].head(20)
```

```python
# 3. Преобразуем Embarked в бинарный признак
df_train_dum = pd.get_dummies(df_train, columns=['Embarked'])
```

```python
df_train_dum.head()
```

```python
# 4. Преобразуем Sex в бинарный признак, где male=1, female=0, удалим атрибут Sex
df_train_dum['Sex_bin'] = [1 if i=='male' else 0 for i in df_train['Sex']]
df_train_dum.drop(columns='Sex', inplace=True, axis=1)
```

```python
df_train_dum.corr()
```

```python
pd.options.display.max_columns = 20000
pd.options.display.max_rows = 20000
pd.options.display.expand_frame_repr = False
X_train = df_train_dum.drop('Survived', axis=1)
y_train = df_train_dum['Survived']
df_test_dum = pd.get_dummies(df_test, columns=['Embarked'])
df_test_dum['Sex_bin'] = [1 if i=='male' else 0 for i in df_test_dum['Sex']]
df_test_dum=df_test_dum.drop(['Sex', 'PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
df_test_dum['Age'].fillna(median_age, inplace=True)
df_test_dum['Fare'].fillna(df_train.Fare.median(), inplace=True)
df_train_dum.to_csv("df2.csv")
df_test_dum
```

```python
print df_test_dum.shape
print X_train.shape
print np.any(np.isnan(df_test_dum))
print np.any(np.isfinite(df_test_dum))
df_test_dum.info()
df_test_dum.reset_index()
```

```python
# модель логистической регрессии
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
pred = model.predict(df_test_dum)
#y_pred=pd.DataFrame(pred)
print df_test.iloc[1]
df_test['Survived']=[i for i in pred]
df_test = df_test.drop(['Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked'], axis=1)
df_test.to_csv("save.csv", index=None)
```
