```python


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
# Функция подготовки датасета для подачи в модель
def make_dataset(df, TRAIN_FLAG):
    str_features=['Alley', 'MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1',\
'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual', \
'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', \
'CentralAir', 'Electrical', 'KitchenQual', 'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',\
'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature', 'SaleType', 'SaleCondition']
    int_features=list(set(df.columns)-set(str_features))
    
    
    # Выделяем признаки и метки
    if TRAIN_FLAG:
        X, y = df.drop(['SalePrice'], axis=1), df['SalePrice']
        int_features.remove('SalePrice')
    else:
        X = df
    print X.shape
    # Заполняем категориальные признаки нулями - там где NaN
    X[str_features] = X[str_features].fillna(0)
    
    # Превращаем категориальные признаки в бинарные через One-Hot-Encoding
    encoder = OneHotEncoder(sparse=True)
    encoded_cat_columns = pd.DataFrame(encoder.fit_transform(X[str_features]))
    
    # Присоединяем к исходному датасету перекодированные признаки
    X_onehot_str = pd.concat([X, encoded_cat_columns], axis=1)
    
    # Удаляем из датасета исходные категориальные признаки
    X_onehot_str.drop(inplace=True, columns=str_features, axis=1)
    print X_onehot_str.shape
    # Заполняем отдельные числовые признаки средним по столбцу
    X_onehot_str['MasVnrArea']=X_onehot_str['MasVnrArea'].fillna(value=np.mean(X.MasVnrArea))
    nan_LotFrontage = np.mean(X.LotFrontage.dropna())
    X_onehot_str['LotFrontage']=X_onehot_str['LotFrontage'].fillna(value=nan_LotFrontage)
    nan_GarageYrBlt = int(np.mean(X.GarageYrBlt.dropna()))
    X_onehot_str['GarageYrBlt']=X_onehot_str['GarageYrBlt'].fillna(value=nan_GarageYrBlt)
    #nan_GarageCars=
    X_onehot_str['GarageCars']=X_onehot_str['GarageCars'].fillna(value=int(np.mean(X.GarageCars.dropna())))
    nan_GarageYrBlt = int(np.mean(X.GarageArea.dropna()))
    X_onehot_str['GarageArea']=X_onehot_str['GarageArea'].fillna(value=nan_GarageYrBlt)
    nan_GarageYrBlt = int(np.mean(X.BsmtFullBath.dropna()))
    X_onehot_str['BsmtFullBath']=X_onehot_str['BsmtFullBath'].fillna(value=nan_GarageYrBlt)
    nan_GarageYrBlt = int(np.mean(X.HalfBath.dropna()))
    X_onehot_str['HalfBath']=X_onehot_str['HalfBath'].fillna(value=nan_GarageYrBlt)
    nan_GarageYrBlt = int(np.mean(X.BsmtHalfBath.dropna()))
    X_onehot_str['BsmtHalfBath']=X_onehot_str['BsmtHalfBath'].fillna(value=nan_GarageYrBlt)
    nan_GarageYrBlt = int(np.mean(X.BsmtFinSF1.dropna()))
    X_onehot_str['BsmtFinSF1']=X_onehot_str['BsmtFinSF1'].fillna(value=nan_GarageYrBlt)
    nan_GarageYrBlt = int(np.mean(X.BsmtFinSF2.dropna()))
    X_onehot_str['BsmtFinSF2']=X_onehot_str['BsmtFinSF2'].fillna(value=nan_GarageYrBlt)
    nan_GarageYrBlt = int(np.mean(X.BsmtUnfSF.dropna()))
    X_onehot_str['BsmtUnfSF']=X_onehot_str['BsmtUnfSF'].fillna(value=nan_GarageYrBlt)
    nan_GarageYrBlt = int(np.mean(X.TotalBsmtSF.dropna()))
    X_onehot_str['TotalBsmtSF']=X_onehot_str['TotalBsmtSF'].fillna(value=nan_GarageYrBlt)
    print X_onehot_str.shape
    if TRAIN_FLAG:
        return X_onehot_str, y
    else:
        return X_onehot_str   
```
